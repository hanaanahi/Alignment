#DOT PLOT AND DYNAMIC PROGRAMMING

```python
import numpy as np
import matplotlib.pyplot as plt

class TextFileReader:
    def __init__(self, filenames):
        self.filenames = filenames

    def read_and_process(self):
        all_sequences = []
        for filename in self.filenames:
            with open(filename, 'r') as file:
                content = file.readlines()
                content.pop(0)  # removes metadata
                cleaned_content = []
                for line in content:
                    cleaned_line = line.strip()  # removes beginning and leading whitespaces
                    cleaned_content.append(cleaned_line)
                sequence = ''.join(cleaned_content)
                sequence_list = list(sequence)
                all_sequences.append(sequence_list)
        return all_sequences

class DotPlot:
    def __init__(self, filenames):
        self.file_reader = TextFileReader(filenames)

    def trim_sequence(self, seq1, seq2):
        if len(seq1) < len(seq2):
            min_len = len(seq1)
        else:
            min_len = len(seq2)
        return seq1[:min_len], seq2[:min_len]

    def create_dot_plot_matrix(self):
        sequences = self.file_reader.read_and_process()
        seq1, seq2 = sequences
        seq1, seq2 = self.trim_sequence(seq1, seq2)  # Trim sequences to the same length
        dot_matrix = np.zeros((len(seq1), len(seq2)))
        for i in range(len(seq1)):
            for j in range(len(seq2)):
                if seq1[i] == seq2[j]:
                    dot_matrix[i, j] = 1
        return dot_matrix

    def plot_dot_plot_matrix(self, dot_matrix, length):
        plt.figure(figsize=(10, 8))
        plt.imshow(dot_matrix, cmap='viridis', interpolation='nearest')
        plt.title('Dot Plot Matrix')
        plt.xlabel('Sequence 1 (Length: {})'.format(length))
        plt.ylabel('Sequence 2 (Length: {})'.format(length))
        plt.colorbar()
        plt.show()

class GlobalAlignment:
    def __init__(self, seq1, seq2, match_score, mismatch_score, gap_penalty):
        self.seq1 = seq1
        self.seq2 = seq2
        self.match_score = match_score
        self.mismatch_score = mismatch_score
        self.gap_penalty = gap_penalty
        self.align_seq1 = ""
        self.align_seq2 = ""
        self.align_marker = ""
        self.align_score = 0

    def align_matrix_calc(self):
        k = len(self.seq1)
        l = len(self.seq2)
        res = []
        for seq in range(k + 1):
            res.append([0] * (l + 1))

        for i in range(1, k + 1):
            res[i][0] = res[i - 1][0] + self.gap_penalty
        for j in range(1, l + 1):
            res[0][j] = res[0][j - 1] + self.gap_penalty

        for i in range(1, k + 1):
            for j in range(1, l + 1):
                if self.seq1[i - 1] == self.seq2[j - 1]:
                    diagonal_box = res[i - 1][j - 1] + self.match_score
                else:
                    diagonal_box = res[i - 1][j - 1] + self.mismatch_score
                top_box = res[i - 1][j] + self.gap_penalty
                beside_box = res[i][j - 1] + self.gap_penalty
                res[i][j] = max(diagonal_box, top_box, beside_box)

        return res

    def align_tracer(self, res):
        p = len(self.seq1)
        q = len(self.seq2)

        while p > 0 or q > 0:
            if p > 0 and q > 0:
                if self.seq1[p - 1] == self.seq2[q - 1]:  # Check if the base pairs match
                    score = self.match_score
                    # self.align_marker = "|" + self.align_marker  # Add "|" if base pairs match
                else:
                    score = self.mismatch_score
                    # self.align_marker = " " + self.align_marker  # Add space if base pairs don't match

                self.align_seq1 = self.seq1[p - 1] + self.align_seq1
                self.align_seq2 = self.seq2[q - 1] + self.align_seq2

                p -= 1
                q -= 1

            elif p > 0 and res[p][q] == res[p - 1][q] + self.gap_penalty:
                self.align_seq1 = self.seq1[p - 1] + self.align_seq1
                self.align_seq2 = "-" + self.align_seq2
                self.align_marker = " " + self.align_marker

                p -= 1

            else:
                self.align_seq1 = "-" + self.align_seq1
                self.align_seq2 = self.seq2[q - 1] + self.align_seq2
                self.align_marker = " " + self.align_marker

                q -= 1

    def align(self):
        align_matrix = self.align_matrix_calc()
        self.align_tracer(align_matrix)
        self.align_score = align_matrix[len(self.seq1)][len(self.seq2)]

# Example usage:
filenames = ['APOE.gene .txt', 'TREM2.txt']
file_reader = TextFileReader(filenames)
sequences = file_reader.read_and_process()
seq1, seq2 = sequences

dot_plotter = DotPlot(filenames)
dot_matrix = dot_plotter.create_dot_plot_matrix()
dot_plotter.plot_dot_plot_matrix(dot_matrix, min(len(seq1), len(seq2)))

match_score = 1
mismatch_score = -1
gap_penalty = -1

# Use trimmed sequences for alignment
trimmed_seq1, trimmed_seq2 = dot_plotter.trim_sequence(seq1, seq2)
alignment = GlobalAlignment(trimmed_seq1, trimmed_seq2, match_score, mismatch_score, gap_penalty)
alignment.align()

print("Alignment Sequence 1:", alignment.align_seq1)
# print("Alignment Marker    :", alignment.align_marker)
print("Alignment Sequence 2:", alignment.align_seq2)
print("Alignment Score:", alignment.align_score)
```
